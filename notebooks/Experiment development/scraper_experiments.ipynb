```python
import numpy as np
import pandas as pd
import re
import os

files = [f for f in os.listdir("../data/01_raw") if re.match(r"anime_scrapy_damien_r[0-9].csv",f)]
print(files)

for file in files:
    df = pd.read_csv(f"../data/01_raw/{file}", sep="|")
    print(len(df))
```

```python
# Concat scraping attempts
df1 = pd.read_csv(f"../data/01_raw/anime_scrapy_damien_r1.csv", sep="|")
df2 = pd.read_csv(f"../data/01_raw/anime_scrapy_damien_r2.csv", sep="|")
files = [df1, df2]
df = pd.concat(files)
print(len(df))
print(f"Number of missing ids = {2800-len(df)}")

# Export scraping attempts
df.to_csv("../data/01_raw/anime_scrapy_damien.csv", sep="|")
```
